# 분할표

```python
import pandas as pd

df = pd.read_csv('chap4_2.1.csv', header=1)
df
```

```python
pd.to_numeric(df['안함'])
pd.to_numeric(df['약하게 함'])
pd.to_numeric(df['보통으로 함'])
pd.to_numeric(df['심하게 함'])
```

```python
rf = []
medicine = ["알약", "가짜약", "합계"]
for i in range(len(df)):
    rf.append([])
    rf[i].append(medicine[i])
    rf[i].append(df['안함'][i]/200)
    rf[i].append(df['약하게 함'][i]/200)
    rf[i].append(df['보통으로 함'][i]/200)
    rf[i].append(df['심하게 함'][i]/200)
    rf[i].append(df['Unnamed: 5'][i]/200)
rf
```

# 상대도수 분할표

```python
level = ['종류', '안함', '약하게 함', '보통으로 함', '심하게 함', '합계']
df_rf = pd.DataFrame(rf, columns=level)
df_rf
```

# 표본상관계수

```python
import pandas as pd

df5 = pd.read_csv('chap4_ex5.csv', header=0)
df5
```

```python
import numpy as np
height = np.array(df5['키(츠)'])
weight = np.array(df5['몸무게(kg)'])
```

```python
np.corrcoef(height,weight) #correlation coefficient
```

```python
np.corrcoef(height,weight)[1][0] #[i][j] 1행, 0열 
```

# 산점도

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.scatter(height, weight, color='black')
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.title('height(cm) and weight(kg)')
plt.show()
```
